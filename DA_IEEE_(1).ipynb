```python
df_numbers = df[["Age","Height","Weight","Year"]]
df_numbers.head()
```

```python
# Distrubution -> hist | outliers -> boxplot | corr -> heatmap

import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for distribution
numeric_cols = df_numbers.columns
df[numeric_cols].hist(bins=15, figsize=(15, 6), layout=(2, 4))
plt.show()

# Boxplot for outliers
plt.figure(figsize=(15, 6))
df[numeric_cols].boxplot()
plt.show()

# Heatmap for correlation
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
```

```python
import matplotlib.pyplot as plt
name_counts = df['NOC'].value_counts()

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(name_counts, labels=name_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Pie Chart of Categories')
plt.show()
```
